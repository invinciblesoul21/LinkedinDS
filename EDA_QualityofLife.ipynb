```python
sns.lineplot(x='Safety Category',y='Health Care Value',color='red',data=df1)
plt.xlabel('Safety Category')
plt.ylabel('Health Care Value')
plt.title('Safety vs Health Care')
plt.show()
```

```python
sns.lineplot(x='Purchasing Power Category',y='Health Care Value',color='green',data=df1)
plt.xlabel('Purchasing Power category')
plt.ylabel('Health Care Value')
plt.title('Purchasing vs Health Care')
plt.show()
```

```python
'''Countries with high purchasing category tends to have higher health care value too.'''
```

```python
'''Q2 Is there a noticeable trend or pattern between the purchasing power category and the quality of life category?'''

cross_tab = pd.crosstab(df1['Quality of Life Category'], df1['Purchasing Power Category'])
plt.figure(figsize=(8, 6))
sns.heatmap(cross_tab, annot=True, cmap='plasma', fmt='d')
plt.title('Relationship between Quality and Purchasing Power')
plt.show()
```

```python
'''Very high purchasing power tends to have high quality of life.'''
```

```python
'''Q3 Are countries with higher safety values also high in other factors like health care or climate?'''



plt.figure(figsize=(15,10))
plt.subplot(2,1,1)
sns.lineplot(x='Health Care Category',y='Safety Value',data=df1,color='cyan',marker='o')
plt.xlabel('Health Care Category')
plt.ylabel('Safety Value')
plt.title('Health vs Safety ')

plt.figure(figsize=(15,10))
plt.subplot(2,1,2)
sns.lineplot(x='Climate Category',y='Safety Value',data=df1,color='brown',marker='o')
plt.xlabel('Climate Category')
plt.ylabel('Safety Value')
plt.title('Climate vs Safety')
```

```python
'''Based on above two plots we can infer that higher health values tends to have higher safety values but moderate climate tends to have higher safety values. We can also infer that countries with countries
with modearte climate means niether too hot nor to cold have better safety facilities but too warm countries usually africans as we know have lower safety safety measures.'''
```

```python
'''Q4 How does safety value correlate with pollution or traffic commute time?'''

c1=df1['Safety Value'].corr(df1['Pollution Value'])
print(c1)
```

```python
c2=df1['Safety Value'].corr(df1['Traffic Commute Time Value'])
print(c2)
```

```python
'''As we know that positive correlation value tends to have direct relationship whereas negative correlation value tends to have inverse relationship.
So, from above observations c1 and c2 are both inversely related but c2 is more inversely related than c1.'''
```
